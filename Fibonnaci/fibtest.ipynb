```python
%matplotlib inline
import time
import numpy as np
import matplotlib.pyplot as plt
```

Implementación de fibonnaci recursivo:

```python
def fibR(n):
    if(n == 0):
        return 0
    elif(n == 1):
        return 1
    else:
        return fibR(n-1) + fibR(n-2)
```

Implementación de fibbonaci iterativo:

```python
def fibI(n):
    x = 0
    y = 1
    for i in range(n):
        x, y = y, x + y
    return x
        
```

```python
#test
print(fibI(8), fibR(8))
```

Prueba de desempeño:

```python
def testAux(fib,n):
    t0 = time.clock()
    fib(n)
    t1 = time.clock()
    return t1-t0
    
def test(fib, n):
    X = np.arange(n)
    Y = [testAux(fib,i) for i in range(n)]
    plt.xlabel('N')
    plt.ylabel('Tiempo(ms)')
    plt.plot(X,Y);
    
```

```python
n = 30
test(fibR,n)
test(fibI,n)
```

```python
test(fibR,n)
plt.show()
test(fibI,n*10)
```
